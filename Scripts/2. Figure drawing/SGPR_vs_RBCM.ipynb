```python
import numpy as np
import matplotlib.pyplot as plt


phases = ["Gas", "Clusters", "Liquid", "Solid"]
metrics = ["Energy", "Forces"]

energy_sgpr = [5.945558, 2.213339, 1.22194, 0.958] 
energy_rbcm = [7.105755, 2.432759, 5.990182, 0.792] 

force_sgpr = [107.919587, 76.503021, 62.769464, 101.07117] 
force_rbcm = [88.025858, 74.216263, 59.822591, 105.411697] 

vib_metrics = ["Frequency", "ZPE"]
frequency_sgpr = [3.145897177]
frequency_rbcm = [2.908988575]

zpe_sgpr = [32.25284738]
zpe_rbcm = [38.57403189]

x = np.arange(len(phases) * len(metrics) + len(vib_metrics))  

fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.spines["top"].set_visible(True)
ax1.spines["right"].set_visible(True)
ax1.spines["bottom"].set_visible(True)
ax1.spines["left"].set_visible(True)

width = 0.3  

bar_energy_sgpr = ax1.bar(x[:len(phases)*2:2] - width/2, energy_sgpr, width, color='blue', label="Energy SGPR")
bar_energy_rbcm = ax1.bar(x[:len(phases)*2:2] + width/2, energy_rbcm, width, color='red', label="Energy RBCM")

bar_frequency_sgpr = ax1.bar(x[-2] - width/2, frequency_sgpr[0], width, color='blue', label="Frequency SGPR")  
bar_frequency_rbcm = ax1.bar(x[-2] + width/2, frequency_rbcm[0], width, color='red', label="Frequency RBCM")  

ax1.set_ylim(0, 15)  

ax2 = ax1.twinx()
bar_force_sgpr = ax2.bar(x[1:len(phases)*2:2] - width/2, force_sgpr, width, color='blue', hatch='//', alpha=1.0, label="Force SGPR")  
bar_force_rbcm = ax2.bar(x[1:len(phases)*2:2] + width/2, force_rbcm, width, color='red', hatch='//', alpha=1.0, label="Force RBCM")  

bar_zpe_sgpr = ax2.bar(x[-1] - width/2, zpe_sgpr[0], width, color='blue', hatch='//', alpha=1.0, label="ZPE SGPR") 
bar_zpe_rbcm = ax2.bar(x[-1] + width/2, zpe_rbcm[0], width, color='red', hatch='//', alpha=1.0, label="ZPE RBCM")  

ax2.set_ylim(0, 150)  

xtick_labels = [f"{metric}\n{phase}" for phase in phases for metric in metrics] + vib_metrics
ax1.set_xticks(x)
ax1.set_xticklabels(xtick_labels)

ax1.set_xlabel("")
ax1.set_ylabel("")
ax2.set_ylabel("")

fig.subplots_adjust(right=0.75)  

legend1 = ax1.legend(
    handles=[bar_energy_sgpr, bar_energy_rbcm, bar_force_sgpr, bar_force_rbcm],
    labels=["Energy SGPR", "Energy RBCM", "Force SGPR", "Force RBCM"],
    loc="upper left",
    bbox_to_anchor=(1.05, 1.0), 
    frameon=True,
)

legend2 = ax2.legend(
    handles=[bar_frequency_sgpr, bar_frequency_rbcm, bar_zpe_sgpr, bar_zpe_rbcm],
    labels=["Frequency SGPR", "Frequency RBCM", "ZPE SGPR", "ZPE RBCM"],
    loc="upper left",
    bbox_to_anchor=(1.05, 0.8), 
    frameon=True,
)

plt.show()
```

```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Categories
categories = ["Alcohol", "Aldehyde", "Carboxylate", "Ester", "Ether", "Sugar", "Lactone", "Olide"]

# RBCM data 
energy_rbcm = [4.34616290762557, 3.15585206289543, 3.11030821986343, 2.80930629796826,
               4.31439623421808, 4.01563185988367, 3.72463880245809, 3.84024899499905]
force_rbcm = [133.426994852672, 129.583155247605, 122.296040417235, 104.628878785597,
              133.190645262997, 137.6728435016, 143.812443761261, 155.965027416807]

# SGPR data 
energy_sgpr = [4.3, 3.1, 2.9, 2.8, 5.2, 4.3, 2.9, 3.9]
force_sgpr = [122.63, 124.31, 107.77, 104.28, 131.15, 147.31, 129.61, 158.38]

# x-position
x = np.arange(len(categories))
bar_width = 0.2  
gap = 0.05  # space of Energy MAE and Force MAE

# Graph
fig, ax1 = plt.subplots(figsize=(10, 6))

# Energy MAE graph bar (left- y axis, Blue/Red)
bar_energy_sgpr = ax1.bar(x - (1.5 * bar_width + gap), energy_sgpr, bar_width, color='blue', label="Energy SGPR", zorder=3)
bar_energy_rbcm = ax1.bar(x - (0.5 * bar_width + gap), energy_rbcm, bar_width, color='red', label="Energy RBCM", zorder=3)

ax1.set_ylabel("Energy MAE (meV/atom)")
ax1.set_ylim(0, 6)  # left y-axis Maximum value

# Force MAE graph bar (right y-axis,Maximum value, Dash Blue/Red)
ax2 = ax1.twinx()
bar_force_sgpr = ax2.bar(x + (0.5 * bar_width + gap), force_sgpr, bar_width, color='blue', hatch='//', alpha=1.0, label="Force SGPR", zorder=2)
bar_force_rbcm = ax2.bar(x + (1.5 * bar_width + gap), force_rbcm, bar_width, color='red', hatch='//', alpha=1.0, label="Force RBCM", zorder=2)

ax2.set_ylabel("Force MAE (meV/)")
ax2.set_ylim(0, 180)  # right y-axis Maximum value
ax2.set_yticks(np.arange(0, 181, 30))  # 30 per 1 gradation

# x-axis setting
ax1.set_xticks(x)
ax1.set_xticklabels(categories, rotation=45, ha='right')

# square design 
ax1.spines["top"].set_visible(True)
ax1.spines["right"].set_visible(True)
ax1.spines["bottom"].set_visible(True)
ax1.spines["left"].set_visible(True)

handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles=handles1 + handles2, labels=labels1 + labels2, loc="upper left", frameon=True)

# Title
plt.title("Energy MAE (SGPR / RBCM) / Force MAE (SGPR / RBCM)")

plt.show()
```
